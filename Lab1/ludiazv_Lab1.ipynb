# Data Collections

## List
A list is an ordered collection of values. A list is mutable, which means you can change a list's value without changing the list itself. Creating a list is simply putting different comma-separated values between square brackets.

````python
my_list = ['Mi nombre es', 'Luis'] #, 1,2,3] # Crea una lista my_list con elementos de 'Mi nombre es' y 'Luis'
print(my_list)  #Imprime la lista my_list, la lista empieza y termina con parentesis cuadrados y se separan los elementos por medio de comas 
````

The values in a list are called "elements". We can access list elements by indexing. Python index starts from 0. So if you have a list of length n, the index of the first element will be 0, and that of the last element will be n − 1. By the way, the length of a list can be obtained by the built-in function len().

````python
my_list = ['Mi nombre es', 'Luis', 15, 7 ,2000] #Crea la lista my_list con 5 elementos ('Mi nombre es', 'Luis', 15, 7 ,2000)
print(len(my_list)) # Imprime la longitud de la lista my_list 
print(my_list[0]) # Imprime el elemento 0 (primer elemento) de la lista 
print(my_list[len(my_list) -1]) # Imprime el último elemento de la lista
````

You can also change the elements in the list by accessing an index and assigning a new value.

````python
my_list = ['Mi nombre es', 'Luis', 15, 7 ,2000] # Crea la lista my_list con 5 elementos ('Mi nombre es', 'Luis', 15, 7 ,2000)
my_list[2] = '20' # Reemplaza el elemento 2 (teniento en cuenta que el primero es elemento 0) por el int 20
print(my_list) # Imprime la lista my_list
````

A list can also be sliced with a colon:

````python
my_list = ['Mi nombre es', 'Luis', 15, 7 ,2000] # Crea la lista my_list con 5 elementos ('Mi nombre es', 'Luis', 15, 7 ,2000)
print(my_list[1:4]) # Imprime desde los elementos con índice 1 al 4 (sin incluir el 4) de la lista my_list
````

The slice starts from the first element indicated, but excludes the last element indicated. Here we select all elements starting from index 1, which refers to the second element:

````python
print(my_list[1:]) # Imprime desde los elementos con índice 1 hasta el último (incluyendo el último) elemento de la lista my_list
````

And all elements up to but excluding index 3:

````python
print(my_list[:3]) # Imprime desde el primer elemento hasta el elemento con indice 3 (sin incluir el 3) de la lista my_list
````

If you wish to add or remove an element from a list, you can use the append() and remove() methods for lists as follows:

````python
my_list = ['Mi nombre es', 'Luis'] # Crea la lista my_list con 2 elementos ('Mi nombre es', 'Luis')
my_list.append('Alfonso') # Añade al final my_list el elemento 'Alfonso'
print(my_list) # Imprime la lista my_list
````

````python
my_list.remove('Luis') # Elimina de la final my_list el elemento 'Luis'
print(my_list) # Imprime la lista my_list
````

# Common String Operations

````python
my_str = 'Hola Mundo' # Declara un string my_str "Hola Mundo"
print(my_str[5:]) # Imprime el String desde el caracter 5 (sin incluirlo) en adelante
````

There are many methods associated with strings. We can use string.count() to count the occurrences of a character in a string, use string.find() to return the index of a specific character, and use string.replace() to replace characters

````python
print('Trabaja desde las ocho de la mañana'.count('a')) # Retorna la cantidad de letras 'a' que hay en la frase previa al count
print('Borra esa carpeta de ahí, que no tiene nada'.find('e')) # Retorna la posición de la primera letra 'e' que hay en la frase previa al find
print('Solo puedes ver al oso'.replace('o','e')) # Reemplaza todas las letras 'o' por la letra 'e' en la frase previa al find y retorna la frase con los cambios
````

The most commonly used method for strings is string.split(). This method will split the string by the indicated character and return a list:

````python
login = 'luisDiaz@algorithms.com/passwd_Example' # Se crea el String user_passwd que; en éste caso, almacena la informacion de un correo y una contraseña
split = login.split('/') # Crea una lista split donde cada elemento es el String separado por el argumento de split.
user = split[0] # Guarda en user el primer elemento de la lista split,  es decir, la parte antes del primer '/' en login (por lo tanto, el correo)
password = split[1] # Guarda en password el segundo elemento de la lista split,  es decir, la parte despues del primer '/' en login (por lo tanto, la contraseña)
print(user, password) #Imprime el valor de user y password
domain = user.split('@')[1] # Guarda en domain la parte despues del '@' en user
print(domain) # Imprime domain
````

# Logical Operations
Like most programming languages, Python has comparison operators:

````python
print(8 == 1) # Devuelve el valor de verdad de 8 es igual a 1, es decir, Falso (False) 
print(2 == 2) # Devuelve el valor de verdad de 2 es igual a 2, es decir, Verdadero (True)
print(8 != 2) # Devuelve el valor de verdad de 8 es diferente a 2, es decir, Verdadero (True)
print(5 >= 5) # Devuelve el valor de verdad de 5 es mayor o igual a 5, es decir, Verdadero (True) 
print(5 >= 6) # Devuelve el valor de verdad de 5 es mayor o igual a 6, es decir, Falso (False) 
print(5 <= 5) # Devuelve el valor de verdad de 5 es menor o igual a 5, es decir, Verdadero (True) 
print(5 <= 6) # Devuelve el valor de verdad de 5 es menor o igual a 6, es decir, Verdadero (True) 
````

Each statement above has a boolean value, which must be either True or False, but not both.

We can combine simple statements P and Q to form complex statements using logical operators:

- The statement "P and Q" is true if both P and Q are true, otherwise it is false.
- The statement "P or Q" is false if both P and Q are false, otherwise it is true.
- The statement "not P" is true if P is false, and vice versa.

````python
print(8 > 5 and 5 > 3) # Devuelve el valor de verdad de 8 es mayor a 5 y 5 es mayor a 3, es decir, Verdadero (True)  
print(8 > 3 and 3 < 5) # Devuelve el valor de verdad de 8 es mayor a 5 y 3 es menor a 5, es decir, Verdadero (True) 
print(2 > 4 or 1 < 8) # Devuelve el valor de verdad de 2 es mayor a 4 o 1 es menor a 8, es decir, Verdadero (True) 
print(8 < 1 and 11 < 2) # Devuelve el valor de verdad de 8 es menor a 1 y 11 es menor a 2, es decir, Falso (False) 
````

When dealing with a very complex logical statement that involves in several statements, we can use brackets to separate and combine them.

````python
print((5 >= 5 or 8 < 3) and (1!=7 and 6>1) and not ( 5 < 7 or 8 < 3))

# Devuelve el valor de verdad de
# 5 es mayor o igual a 5 o 8 es menor a 3 y
# 1 es diferente a 7 y 6 mayor que 1 y
# negacion de 5 es menor a 7 u 8 es menor a 3,
# en éste caso, falso (False)

print(8 > 3 or 5 < 11 and (1!=3 and 8>9) and not ( 6 < 2 or 1 < 10))

# Devuelve el valor de verdad de
# 8 es mayor a 3 o 5 es menor a 11 y
# 1 es diferente a 3 y 8 mayor que 9 y
# negacion de 6 es menor a 2 u 1 es menor a 10,
# en éste caso, verdadero (True)
````

Comparing the above two statements, we can see that it's wise to use brackets when we make a complex logical statement.

# If Statement
An if statement executes a segment of code only if its condition is true. A standard if statement consists of 3 segments: if, elif and else.

```python
if statement1:
    # if the statement1 is true, execute the code here.
    # code.....
    # code.....
elif statement2:
    # if the statement 1 is false, skip the codes above to this part.
    # code......
    # code......
else:
    # if none of the above statements is True, skip to this part
    # code......
```

An if statement doesn't necessarily has elif and else part. If it's not specified, the indented block of code will be executed when the condition is true, otherwise the whole if statement will be skipped.

````python
i = 5
if i == 0: # Si i es igual a 0 se realiza el codigo identado debajo de la condicion
    print('i==0 is True') # Imprime 'i==0 is True'
else: # En caso que en las condiciones del if (o los elif) 
    print('i==0 is False') # Imprime 'i==0 is False'
````

As we mentioned above, we can write some complex statements here:

````python
x = 5 > 8   # La variable x guarda el valor de verdad de 5 es mayor a 8, en éste caso, es falso (False)
y = 1 < 7   # La variable y guarda el valor de verdad de 1 es menor a 7, en éste caso, es verdadero (True)

if x and y:   # Si i es igual a 0 se realiza el codigo identado debajo de la condicion
    print('x and y is true')   # Imprime 'x and y is true'

elif x and not y:   # En caso de que la condicion del if sea falsa, se verifica la condición del elif 
                    # y si es verdadera se realiza el codigo identado debajo
    print('y is false')   # Imprime 'y is false'

elif y and not x:   # En caso de que la condicion del if y del anterior (o anteriores si fuera el caso) elif sean falsa,
                    # se verifica la condición del siguiente elif y si es verdadera se realiza el codigo identado debajo
    print('x is false')   # Imprime 'x is false'

else:   # En caso de que la condicion del if y de todos los elif sean falsa, se realiza el codigo identado debajo de la condicion
    print('x and y are false')   # Imprime 'x and y are false'
````

# Loop Structure
Loops are an essential part of programming. The "for" and "while" loops run a block of code repeatedly.

## While Loop
A "while" loop will run repeatedly until a certain condition has been met.

````python
i = 0 # Guarda en la variable i el entero 0
while i < 5: # Si i es menor a 5 se realiza el codigo identado debajo siempre en cuando se cumpla la condición
    print(i*2) # Imprime el valor de i multiplicado por 2
    i += 1 # Incrementa el valor de i en 1
````

When making a while loop, we need to ensure that something changes from iteration to iteration so that the while loop will terminate, otherwise, it will run forever. Here we used i += 1 (short for i = i + 1) to make i larger after each iteration. This is the most commonly used method to control a while loop.

## For Loop
A "for" loop will iterate over a sequence of value and terminate when the sequence has ended.

````python
for i in [4,5,6,7,8]: #Para cada elemento (denotado por i) en la lista ([4,5,6,7,8]) se realiza el codigo identado debajo
    print(i%2) #Imprime el residuo de i dividido entre 2
````

We can also add if statements in a for loop. Here is a real example from our pairs trading algorithm:

````python
chars = ['A','B','C','D','E','F','G','H'] # Se crea una lista char con elementos 'A','B','C','D','E','F','G' y 'H'
removeChars = ['A','D','F'] #Se crea una lista removeChars con elementos 'A','D' y 'F'
finalList = chars # Se crea una lista finalList que es una copia de chars
for i in chars: # Para cada elemento i en la lista chars se realizará el código identado debajo
    if i in removeChars: # Si el elemnto i (iterado de chars) está en la lista removeChars se realizará el código identado debajo
        finalList.remove(i) #Elimina de finalList el elemento i (iterado de chars y perteneciente a la lista removeChars)
print(finalList) #Imprime la lista finalList
````

Here we iterated all the elements in the list 'stocks'. Later in this chapter, we will introduce a smarter way to do this, which is just a one-line code.

## Break and continue
These are two commonly used commands in a for loop. If "break" is triggered while a loop is executing, the loop will terminate immediately:

````python
chars = ['A','B','C','D','E','F','G','H'] # Se crea una lista char con elementos 'A','B','C','D','E','F','G' y 'H'
for i in chars: # Para cada elemento i en la lista chars se realizará el código identado debajo 
    print(i) # Imprime el elemento i (iterado de chars)
    if i == 'E': # Si el elemnto i (iterado de chars) es igual a 'E' se realizará el código identado debajo
        break # Detiene todas las siguientes iteraciones del for
````

The "continue" command tells the loop to end this iteration and skip to the next iteration:

````python
chars = ['A','B','C','D','E','F','G','H'] # Se crea una lista char con elementos 'A','B','C','D','E','F','G' y 'H'
for i in chars: # Para cada elemento i en la lista chars se realizará el código identado debajo 
    if i == 'E':  # Si el elemnto i (iterado de chars) es igual a 'E' se realizará el código identado debajo
        continue # Continua directamente a la siguiente iteración
    print(i) # Imprime el elemento i (iterado de chars)
````

# List Comprehension
List comprehension is a Pythonic way to create lists. Common applications are to make new lists where each element is the result of some operations applied to each member of another sequence. For example, if we want to create a list of squares using for loop:

````python
cuadruple = [] # Se crea una lista cuadruple sin elementos
for i in [0,2,4,6,8]:  # Para cada elemento i en la lista [1,2,3,4,5] se realizará el código identado debajo 
    cuadruple.append(i*4) #Se añade a la lista cuadruple el valor de cada i multiplicado por 4
print(cuadruple) # Imprime la lista cuadruple
````

Using list comprehension:

````python
evenNumbers = [0,2,4,6,8]  # Se crea una evenNumbers cuadruple con elementos 0, 2, 4, 6 y 8
cubes = [x**2 for x in evenNumbers] # Se crea una cubes cuyos elementos son el cubo de cada elemento x (iterado de la lista evenNumbers)
print(cubes) # Imprime la lista cubes
````

Recall the example above where we used a for loop to select stocks. Here we use list comprehension:

````python
chars = ['A','B','C','D','E','F','G','H'] # Se crea una lista char con elementos 'A','B','C','D','E','F','G' y 'H'
removeChars = ['A','D','F'] #S e crea una lista removeChars con elementos 'A','D' y 'F'
finalList = [x for x in chars if x not in removeChars] # Se crea una lista  cuyos cada elemento x (iterado de la lista chars) que no estén en la lista removeChars
print(finalList) # Imprime la lista finalList
````

A list comprehension consists of square brackets containing an expression followed by a "for" clause, and possibly "for" or "if" clauses. For example:

````python
print([(x, y) for x in [1,3,5] for y in [5,7,9] if x != y]) # Imprime las duplas tomando elementos de cada una de las listas tal que los valores de las duplas sean distintos 
print([str(x)+' and '+str(y) for x in ['A','C','D','E'] for y in ['C','D','E','F'] if x!=y]) # Imprime las duplas tomando elementos de cada una de las listas tal que los valores de las duplas sean distintos 
````

List comprehension is an elegant way to organize one or more for loops when creating a list.

````python
list(map(lambda x, y: x+y, [6,7,8,9,10],[5,4,3,2,1]))# Genera una lista utilizando un map, de manera que los elementos de la lista generada son la suma de los
                                                     # elementos x de la primera lista, y elementos y de la segunda lista, que están en la misma posición
````

# Object-Oriented Programming
Python is an object-oriented programming language. It's important to understand the concept of "objects" because almost every kind of data from QuantConnect API is an object.

## Class
A class is a type of data, just like a string, float, or list. When we create an object of that data type, we call it an instance of a class.

In Python, everything is an object - everything is an instance of some class. The data stored inside an object are called attributes, and the functions which are associated with the object are called methods.

For example, as mentioned above, a list is an object of the "list" class, and it has a method list.sort().

We can create our own objects by defining a class. We would do this when it's helpful to group certain functions together. For example, we define a class named "Stock" here:

````python
class player: # Declaración de la clase player
    
    def __init__(self, name, lastname, club, teamNumber): # Método constructor (inicializa la clase)
        self.name = name                                  # Se la pasan los atributos y se los asignan a cada nuevo objeto que se cree
        self.lastname = lastname
        self.club = club
        self.teamNumber = teamNumber
 
    def update(self, name, lastname, club, teamNumber): #Reescribe los atributos del objeto player
        self.name = name                                  
        self.lastname = lastname
        self.club = club
        self.teamNumber = teamNumber
 
    def printPlayer(self): # Imprime toda la información de atributos del objeto player
        print("--------")
        print(self.name)
        print(self.lastname)
        print(self.club)
        print(self.teamNumber)
        print("--------")
````

The "Stock" class has attributes "ticker", "open", "close", "volume" and "rate_return". Inside the class body, the first method is called __init__, which is a special method. When we create a new instance of the class, the __init__ method is immediately executed with all the parameters that we pass to the "Stock" object. The purpose of this method is to set up a new "Stock" object using data we have provided.

Here we create two Stock objects named "apple" and "google".

````python
P1 = player('Mohamed', 'Salah', 'Liverpool', 11) # Crea un objeto player P1 con atributos 'Mohamed', 'Salah', 'Liverpool' y 11
P2 = player('Kai', 'Havertz', 'Chelsea', 29) # Crea un objeto player P2 con atributos 'Kai', 'Havertz', 'Chelsea' y 29
````

Stock objects also have two other methods: update() and print_return(). We can access the attribues of a Stock object and call its methods:

````python
P1.lastname # Devuelve el atributo lastname en el objeto P1
P2.printPlayer() # Imprime toda la información de atributos del objeto P2 
P2.update('Kai', 'Havertz','Bayern Munich',29) #Reescribe los atributos de club por 'Bayern Munich' y teamNumber por '29' al objeto P2
P2.printPlayer() # Imprime toda la información de atributos del objeto P2
````

By calling the update() function, we updated the open and close prices of a stock. Please note that when we use the attributes or call the methods **inside a class**, we need to specify them as self.attribute or self.method(), otherwise Python will deem them as global variables and thus raise an error.

We can add an attribute to an object anywhere:

````python
P2.club = 'Bayer Leverkusen' # Reescribe el atributo club por 'Bayer Leverkusen' al objeto P2
P2.club # Devuelve el atributo club en el objeto P2 
````

We can check what names (i.e. attributes and methods) are defined on an object using the dir() function:

````python
dir(P1) #Lista los atributos y métodos definidos en el objeto P1 (objeto de la clase Player)
````

## Inheritance
Inheritance is a way of arranging classes in a hierarchy from the most general to the most specific. A "child" class is a more specific type of a "parent" class because a child class will inherit all the attribues and methods of its parent. For example, we define a class named "Child" which inherits "Stock":

````python
class newPlayerClass(player):
    def __init__(self,previousTeam):
        self.previousTeam = previousTeam
````

As seen above, the new class Child has inherited the methods from Stock.

# Resampling
**_series.resample(freq)_** is a class called "DatetimeIndexResampler" which groups data in a Series object into regular time intervals. The argument "freq" determines the length of each interval.

**_series.resample.mean()_** is a complete statement that groups data into intervals, and then compute the mean of each interval. For example, if we want to aggregate the daily data into monthly data by mean:

````python
monthlyClose = Nvidia.resample('M').mean() # Agrupa los datos en meses (mediante el resample('M')) y calcula la media del precio de cierre de la acción de Nvidia para cada mes
print(monthlyClose) # Imprime la nueva tabla agrupada por meses
````

We can also aggregate the data by week:

````python
weeklyClose = Nvidia.resample('W').mean() # Agrupa los datos en semanas (mediante el resample('M')) y calcula la media del precio de cierre de la acción de Nvidia para cada semana
print(weeklyClose.head()) # Imprime la nueva tabla agrupada por semanas identificandolas como el primer día de cada semana (domingo de cada semana)
````

We can also aggregate the data by month with max:

````python
Nvidia.resample('M').max() # Agrupa los datos en semanas (mediante el resample('M')) y muestra el precio máximo de cierre de la acción de Nvidia que tuvo durante cada mes
````

We can choose almost any frequency by using the format 'nf', where 'n' is an integer and 'f' is M for month, W for week and D for day.

````python
threeDay = Nvidia.resample('3D').mean() # Agrupa los datos por cada 3 días (mediante el resample('3D')) y calcula la media del precio de cierre de la acción de Nvidia para cada 3 días
twoWeek = Nvidia.resample('2W').mean() # Agrupa los datos por cada 2 semanas (mediante el resample('2W')) y calcula la media del precio de cierre de la acción de Nvidia para cada 2 semanas
twoMonth = Nvidia.resample('2M').mean() # Agrupa los datos por cada 2 meses (mediante el resample('2M')) y calcula la media del precio de cierre de la acción de Nvidia para cada 2 meses


print(threeDay) # Imprime la tabla agrupada por cada 3 días
print(twoWeek) # Imprime la tabla agrupada por cada 2 semanas
print(twoMonth ) # Imprime la tabla agrupada por cada 2 meses
````

Besides the mean() method, other methods can also be used with the resampler:



````python
std = Nvidia.resample('W').std() # Agrupa los datos por cada semana  y calcula la desviación estandar del precio de cierre de la acción de Nvidia para cada semana
max = Nvidia.resample('W').max() # Agrupa los datos por cada semana  y guarda el precio máximo de cierre de la acción de Nvidia para cada semana
min = Nvidia.resample('W').min() # Agrupa los datos por cada semana  y guarda el precio mínimo de cierre de la acción de Nvidia para cada semana


print(std) # Imprime la desviación estandar del precio de cierre de la acción de Nvidia para cada semana
print(max) # Imprime el precio máximo de cierre de la acción de Nvidia para cada semana
print(min) # Imprime el precio mínimo de cierre de la acción de Nvidia para cada semana
````

Often we want to calculate monthly returns of a stock, based on prices on the last day of each month. To fetch those prices, we use the series.resample.agg() method:

````python
lastDay = Nvidia.resample('M').agg(lambda x: x[-1]) # lastDay guarda el último valor de cierre de cada mes 
print(lastDay) # Imprime los datos de lastDay
````

Or directly calculate the monthly rates of return using the data for the first day and the last day:

````python
monthlyReturn = Nvidia.resample('M').agg(lambda x: x[-1]/x[0] - 1) # monthlyReturn guarda a ganancia obtenida respecto al precio de compra de la acción mensualmente
print(monthlyReturn) # Imprime la ganancia obtenida respecto al precio de compra de la acción mensualmente
````

Series object also provides us some convenient methods to do some quick calculation.

````python
print(monthlyReturn.mean()) # Imprime el promedio de ganancia obtenida respecto al precio de compra de la acción mensualmente
print(monthlyReturn.std()) # Imprime la desviación estandar de ganancia obtenida respecto al precio de compra de la acción mensualmente
print(monthlyReturn.max()) # Imprime la maxima ganancia obtenida respecto al precio de compra de la acción mensualmente
````

Another two methods frequently used on Series are .diff() and .pct_change(). The former calculates the difference between consecutive elements, and the latter calculates the percentage change.

````python
print(lastDay.diff()) # Diferencia del precio de la acción de Nvidia del ultimo día del mes y del último dia del mes anterior
print(lastDay.pct_change()) # Porcentaje del cambio del precio de la acción de Nvidia del ultimo día del mes y del último dia del mes anterior
````

Notice that we induced a NaN value while calculating percentage changes i.e. returns.

When dealing with NaN values, we usually either removing the data point or fill it with a specific value. Here we fill it with 0:

````python
dailyReturn = lastDay.pct_change() # dailyReturn guarda el porcentaje del cambio del precio de la acción de Nvidia del ultimo día del mes y del último dia del mes anterior
print(dailyReturn.fillna(0)) # Imprime dailyReturn y cuyos elementos sean inexistentes (NaN), lo reemplaza por 0
````

Alternatively, we can fill a NaN with the next fitted value. This is called 'backward fill', or 'bfill' in short:

````python
dailyReturn = lastDay.pct_change() # dailyReturn guarda el porcentaje del cambio del precio de la acción de Nvidia del ultimo día del mes y del último dia del mes anterior
print(dailyReturn.fillna(method = 'bfill')) # Imprime dailyReturn y cuyos elementos sean inexistentes (NaN), lo reemplaza por el mismo porcentaje del cambio del precio de la acción del último día de cada mes
````

As expected, since there is a 'backward fill' method, there must be a 'forward fill' method, or 'ffill' in short. However we can't use it here because the NaN is the first value.

We can also simply remove NaN values by **_.dropna()_**

````python
dailyReturn = lastDay.pct_change() # # dailyReturn guarda el porcentaje del cambio del precio de la acción de Nvidia del ultimo día del mes y del último dia del mes anterior
dailyReturn.dropna() # # Imprime dailyReturn y cuyos elementos sean inexistentes (NaN), los elimina del registo
````

## DataFrame Concat
We have seen how to extract a Series from a dataFrame. Now we need to consider how to merge a Series or a DataFrame into another one.

In Pandas, the function **concat()** allows us to merge multiple Series into a DataFrame:

````python
A = pd.Series([143.5, 144.09, 142.73, 144.18, 143.77], name = 'AMD') # Crea una serie A con elementos 143.5, 144.09, 142.73, 144.18 y 143.77 , y nombre 'AMD'
B = pd.Series([898.7, 911.71, 906.69, 918.59, 926.99], name = 'NVIDIA') # Crea una serie B con elementos 898.7, 911.71, 906.69, 918.59 y 926.99 , y nombre 'NVIDIA'
DFrame = pd.concat([A,B], axis = 1) # Une la serie A y la serie B en un dataframe con un sólo identificador y lo guarda en DFrame
print(DFrame) # Imprime DFrame
````

The "axis = 1" parameter will join two DataFrames by columns:

````python
logPrice = np.log(nvidiaInfo.Close) # Crea un dataframe logPrice que guarda el logaritmo natural del precio de cierres del precio de las acciones de Nvidia
logPrice.name = 'logPrice' # Nombra al dataframe logPrice como 'logPrice'
print(logPrice) # Imprime el dataframe logPrice
print('\n---------------------- separate line--------------------\n') # Separador de impresiones
union = pd.concat([nvidiaInfo, logPrice], axis = 1) # Guarda en 'union' la concatenación de los dataframe nvidiaInfo y logPrice
print(union) # Imprime el dataframe union
````

We can also join two DataFrames by rows. Consider these two DataFrames:

````python
DFrameVolume = nvidiaTable.loc['2022-02':'2022-03',['Volume', 'Stock Splits']].resample('M').agg(lambda x: x[-1]) # Guarda en DFrameVolume la información de febrero de 2022 y marzo de 2022 con la información de 'Volume' y 'Stock Splits' agrupados por mes
print(DFrameVolume) # Imprime el dataframe DFrameVolume
print('\n---------------------- separate line--------------------\n') # Separador de impresiones
DFrameNVDATable= nvidiaTable.loc['2022-02':'2022-03',['Open', 'High', 'Low', 'Close']].resample('M').agg(lambda x: x[-1]) # Guarda en DFrameNVDATable la información de febrero de 2022 y marzo de 2022 con la información de 'Open', 'High', 'Low' y 'Close' agrupados por mes
print(DFrameNVDATable) # Imprime el dataframe DFrameNVDATable
````

Now we merge the DataFrames with our DataFrame 'aapl_bar'

````python
concat = pd.concat([nvidiaInfo, DFrameVolume], axis = 1) # Guarda en 'concat' la concatenación de los dataframe nvidiaInfo y DFrameVolume
print(concat) # Imprime el dataframe concat
````

By default the DataFrame are joined with all of the data. This default options results in zero information loss. We can also merge them by intersection, this is called 'inner join

````python
concat = pd.concat([nvidiaInfo,DFrameVolume],axis = 1, join = 'inner') # Se guarda en 'concat' la concatenacion  de nvidiaInfo y DFrameVolume que tengan en común las fechas
                                                                            # (es decir, si hay una fecha en uno de los dataframes y en el otro no, no se añade a concat)
print(concat) # Imprime el dataframe 'concat'
````

Only the intersection part was left if use 'inner join' method. Now let's try to append a DataFrame to another one:

````python
append = nvidiaInfo.append(DFrameNVDATable) # Añade al dataframe nvidiaInfo las columnas del dataframe DFrameNVDATable, y se guarda en el dataframe 'append' (usando append)
print(append) # Imprime el dataframe append
````

'Append' is essentially to concat two DataFrames by axis = 0, thus here is an alternative way to append:

````python
noAppend = pd.concat([nvidiaInfo, DFrameNVDATable], axis = 0) # Añade al dataframe nvidiaInfo las columnas del dataframe DFrameNVDATable, y se guarda en el dataframe 'noAppend' (sin usar append)
print(noAppend) # Imprime el dataframe noAppend
````

Please note that if the two DataFrame have some columns with the same column names, these columns are considered to be the same and will be merged. It's very important to have the right column names. If we change a column names here:

````python
DFrameNVDATable.columns = ['Change', 'High','Low','Close'] # Filtra las columnas de DFrameNVDATable por 'Change', 'High', 'Low' y 'Close'
concat = pd.concat([nvidiaInfo, DFrameNVDATable], axis = 0) # Guarda en concat la concatenación de los dataframe nvidiaInfo y DFrameNVDATable
print(concat) # Imprime el dataframe concat
````

Since the column name of 'Open' has been changed, the new DataFrame has an new column named 'Change'.
